```python
import pandas as pd
```

### PERFORMANCE ON 1 MACHINE

```python
one_machine=pd.read_csv("1_machine_times.txt",delimiter="\t",names=["Algo and Size", "Time"])
one_machine
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(one_machine["Algo and Size"],one_machine["Time"])
plt.show()
```
